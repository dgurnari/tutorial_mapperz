```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

import seaborn as sns
from matplotlib import colormaps as cm
```

```python
png_pil_img = Image.open("img/jet.png")
```

```python
png_pil_img
```

```python
png_np_img = np.asarray(png_pil_img.convert("L"))
png_np_img.shape
```

```python
png_np_img
```

```python
points = []

for i in range(png_np_img.shape[0]):
    for j in range(png_np_img.shape[1]):
        if png_np_img[i, j] != 255:
            points.append([j, i])

points = np.array(points)
```

```python
plt.figure(figsize=(30, 5))
sns.scatterplot(
    x=points[:, 0],
    y=points[:, 1][::-1],
    hue=points[:, 1][::-1],
    palette=cm.get_cmap("jet"),
)
plt.xlabel("fuck seaborn")
plt.show()
```

```python
plt.figure(figsize=(30, 5))
sns.scatterplot(
    x=points[:, 0], y=points[:, 1][::-1], hue=points[:, 1], palette=cm.get_cmap("jet")
)
plt.xlabel("fuck seaborn")
plt.show()
```

```python
plt.figure(figsize=(30, 5))
plt.scatter(x=points[:, 0], y=points[:, 1])
plt.xlabel("fuck seaborn")
plt.show()
```
