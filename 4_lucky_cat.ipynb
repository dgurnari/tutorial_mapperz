```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
lucky = pd.read_csv("data/lucky_cat.csv", header=None)
```

```python
X = lucky.to_numpy()
X.shape
```

```python
fig, axs = plt.subplots(ncols=10, figsize=(15, 4))
for i in range(10):
    axs[i].imshow(X[i * 7].reshape((128, 128)), cmap="gray")
    axs[i].set_axis_off()
```
